```python
indexed_dir = r'C:\Users\habib\Documents\GitHub\DataBeach\indexed_df_points'
import os
initial_string = str('Service_side,Start_frame,End_frame')
new_string = ''.join(ch.lower() if ch.isupper() else ch for ch in initial_string)
# print(new_string)

for file_name in os.listdir(indexed_dir):
    if file_name.lower().endswith('.csv'):
        file_path = os.path.join(indexed_dir, file_name)

        with open(file_path, 'r', encoding='utf-8', newline='') as f:
            lines = f.readlines()
        if lines:
            header = lines[0]
            if initial_string in header:
                lines[0] = header.replace(initial_string, new_string)
                with open(file_path, 'w', encoding='utf-8', newline='') as f:
                    f.writelines(lines)
```
