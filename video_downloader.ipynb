```python
import cv2
import pandas as pd
import time

# Chemin vers le fichier vid√©o  

video_path = r'C:\Users\habib\Desktop\Montages volley et beach\Jade&Math\barrage_leuven.mp4'
output_csv = 'stats_beach_volley.csv'
```

```python
# Liste pour stocker les actions
actions = []

# Dictionnaire des touches -> action
key_action_map = {
    ord('a'): 'R√©ception r√©ussie',
    ord('z'): 'R√©ception rat√©e',
    ord('s'): 'Service r√©ussi',
    ord('d'): 'Service rat√©',
}

# Ouvrir la vid√©o
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Erreur : impossible d‚Äôouvrir la vid√©o.")
    exit()

fps = cap.get(cv2.CAP_PROP_FPS)
frame_number = 0

print("\n--- LECTURE EN COURS ---\nAppuie sur une touche pour taguer une action : s/a/f/b/p (ou q pour quitter)\n")

paused = False
last_action = ""  # Variable pour stocker la derni√®re action

while cap.isOpened():
    if not paused:
        ret, frame = cap.read()
        if not ret:
            break
        frame_number += 1

    # Affiche la derni√®re action
    if last_action:
        cv2.putText(frame,
                    f"Derniere action : {last_action}",
                    (30, 40),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    1,
                    (0, 255, 0),
                    2,
                    cv2.LINE_AA)

    # Indiquer mode pause
    if paused:
        cv2.putText(frame,
                    "|| PAUSE ||",
                    (30, 80),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    1,
                    (0, 0, 255),
                    2,
                    cv2.LINE_AA)

    cv2.imshow('Beach Volley Video', frame)

    key = cv2.waitKey(30) & 0xFF

    if key == ord('q'):
        break
    elif key == ord(' '):  # barre espace
        paused = not paused
    elif key in key_action_map:
        timestamp = frame_number / fps
        action_name = key_action_map[key]
        last_action = action_name

        actions.append({
            'Time (s)': round(timestamp, 2),
            'Frame': frame_number,
            'Action': action_name
        })


# last_action = ""
# 
# while cap.isOpened():
#     ret, frame = cap.read()
#     if not ret:
#         break

#     # Affichage du texte √† l‚Äô√©cran
#     if last_action:
#         cv2.putText(frame,
#                     f"Derniere action : {last_action}",
#                     (30, 40),
#                     cv2.FONT_HERSHEY_SIMPLEX,
#                     1,
#                     (0, 255, 0),
#                     2,
#                     cv2.LINE_AA)

#     cv2.imshow('Beach Volley Video', frame)
#     key = cv2.waitKey(30) & 0xFF

#     if key == ord('q'):
#         break
#     elif key in key_action_map:
#         timestamp = frame_number / fps
#         action_name = key_action_map[key]
#         last_action = action_name  # Met √† jour l'action affich√©e

#         actions.append({
#             'Time (s)': round(timestamp, 2),
#             'Frame': frame_number,
#             'Action': action_name
#         })

#         print(f"[{round(timestamp, 2)}s] Action enregistr√©e : {action_name}")

#     frame_number += 1

# Lib√©rer les ressources

cap.release()
cv2.destroyAllWindows()

# Sauvegarder les actions dans un CSV
df = pd.DataFrame(actions)
df.to_csv(output_csv, index=False)
print(f"\n‚úÖ Fichier CSV sauvegard√© : {output_csv}")
```
