```python
import cv2
import pandas as pd
import time

# Chemin vers le fichier vidéo  

video_path = r'C:\Users\habib\Desktop\Montages volley et beach\Martin&Thib vs Lolo&Robin - S2-250 Pont - 14juin2025.mp4'
output_csv = 'stats_beach_volley.csv'
```

```python
# Liste pour stocker les actions
actions = []

# Dictionnaire des touches -> action
key_action_map = {
    ord('a'): 'Réception réussie',
    ord('z'): 'Réception ratée',
    ord('s'): 'Service réussi',
    ord('d'): 'Service raté',
}

# Ouvrir la vidéo
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Erreur : impossible d’ouvrir la vidéo.")
    exit()

fps = cap.get(cv2.CAP_PROP_FPS)
frame_number = 0

print("\n--- LECTURE EN COURS ---\nAppuie sur une touche pour taguer une action : s/a/f/b/p (ou q pour quitter)\n")

paused = False
last_action = ""  # Variable pour stocker la dernière action

try:
    while cap.isOpened():
        if not paused:
            ret, frame = cap.read()
            if not ret:
                print("Fin de la vidéo ou erreur de lecture.")
                break
            frame_number += 1

        # Affiche la dernière action
        if last_action and ret:
            cv2.putText(frame,
                        f"Derniere action : {last_action}",
                        (30, 40),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        1,
                        (0, 255, 0),
                        2,
                        cv2.LINE_AA)

        # Indiquer mode pause
        if paused and ret:
            cv2.putText(frame,
                        "|| PAUSE ||",
                        (30, 80),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        1,
                        (0, 0, 255),
                        2,
                        cv2.LINE_AA)

        if ret:
            cv2.imshow('Beach Volley Video', frame)

        key = cv2.waitKey(30) & 0xFF

        if key == ord('q'):
            print("Quitter.")
            break
        elif key == ord(' '):  # barre espace
            paused = not paused
        elif key in key_action_map:
            timestamp = frame_number / fps
            action_name = key_action_map[key]
            last_action = action_name

            actions.append({
                'Time (s)': round(timestamp, 2),
                'Frame': frame_number,
                'Action': action_name
            })

finally:
    # Libérer les ressources
    cap.release()
    cv2.destroyAllWindows()

# Sauvegarder les actions dans un CSV
if actions:
    df = pd.DataFrame(actions)
    df.to_csv(output_csv, index=False)
    print(f"\n✅ Fichier CSV sauvegardé : {output_csv}")
else:
    print("\n⚠️ Aucune action enregistrée.")
```

```python
# Script pour télécharger une vidéo Facebook en utilisant yt-dlp


def download_facebook_video(dict_liens:dict):

    for video_url, nom_fichier in dict_liens.items():
        ydl_opts = {
            'outtmpl': r'C:/Users/habib/Downloads/%(title)s.%(ext)s',  # Chemin de sortie
            'format': 'best',  # Télécharge le meilleur format disponible sans fusion
            'noplaylist': True,  # Évite de télécharger une playlist
        }

        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            try:
                ydl_opts['cookiefile'] = 'cookies_alexis.txt'
                ydl.params.update(ydl_opts)
                ydl.download([video_url])

                # Fichier téléchargé
                import os, subprocess
                downloaded_file = ydl.prepare_filename(ydl.extract_info(video_url, download=False))

                output_file = f'C:/Users/habib/Downloads/{nom_fichier}.mp4'
                cmd = [
                    "ffmpeg", "-y", "-i", downloaded_file,
                    "-vf", "transpose=2",
                    output_file
                ]
                subprocess.run(cmd, check=True)
                os.remove(downloaded_file)


            except Exception as e:
                print(f"Une erreur s'est produite : {e}")

download_facebook_video(S2_250_Dijon_novembre_2025)
```
